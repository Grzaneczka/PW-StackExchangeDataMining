## Load and extract data for chart race

```python
boardgames = StackExchangeData.load('data/boardgames.pickle')
```

```python
race_data = (
    boardgames.posts
        .assign(Tag = boardgames.posts.Tags.str.strip('<>').str.split('><'))
        .assign(Date = boardgames.posts.CreationDate.dt.date)
        .loc[boardgames.posts.PostTypeId == 1]
        .explode('Tag')
        .groupby(['Date', 'Tag'])
        .size()
        .to_frame('Count')
        .reset_index()
        .pivot('Date', 'Tag', 'Count')
        .fillna(0)
        .astype('int')
        .sort_index()
        .cumsum()
)
```

```python
race_data
```

## Create video

```python
reduced_data = race_data.sort_values(race_data.last_valid_index(), axis=1, ascending=False).iloc[:100, :10]
```

```python
bcr.bar_chart_race(df=reduced_data, n_bars=6, filename=None, period_length=100)
```
